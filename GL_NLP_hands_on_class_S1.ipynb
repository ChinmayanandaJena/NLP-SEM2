```python
all_tokens = []
for doc in docs:
    doc_new = doc.lower()
    tokens = doc_new.split()  # Split the document into tokens (words)
    all_tokens.extend(tokens)

len_all_tokens = len(all_tokens)
print(len_all_tokens)
```

```python
hashtags_tokens = []
for token in all_tokens:
    if token.startswith('#'):
        hashtags_tokens.append(token)

print(hashtags_tokens)
len(hashtags_tokens)

#len_all_tokens = len(all_tokens)
#print(len_all_tokens)
```
